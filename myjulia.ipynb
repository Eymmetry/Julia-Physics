# Markdownの使い方

## 見出し
#一つでh1タグ、##２つでh2タグとなり6つまで記述できます。
## 段落と改行
　段落は空行で一行開ける。改行は行末で半角スペース２つ以上入力します。  
これで改行

これで段落
## 強調
強調したい文字を[*]で囲む
*こんな感じで斜体になる*  
より強調したい場合には[**]で囲む
**こんな感じでblock体になる** 
## リスト 
ハイフン、プラス、アスタリスクのいずれかと半角スペースで箇条書きリストになる。タブか半角スペース２つ分インデントの必要あり
- list1
  - nest
    - nestnest
- list2
  - nest

もしくはハイフンの代わりに1.みたいに書けば番号付きリストになる。こっちはなぜかvscodeだとスペース3つ必要らしい
1. list1
   1. nest
       1. nestnest
1. list2
   1. nest

## 引用
先頭に[>]と半角スペースを入力すると引用になる。入れ子にすることもできる。
> 引用文章
> 引用文章2
>
>>　入れ子にできるよ

## 水平線
[_], [*]を３つ以上並べると水平線を表す
___

## ソースコード
1行だけの場合はバッククォートで囲みます。
複数行の場合は先頭にスペースを4つ入れます。
利用しているエディタによりますが、プログラミング言語ごとのシンタックスハイライトに対応している場合は
「言語名」と「」でソースコードを囲むとハイライトされます。
`int x = 1`
